```python
import pandas as pd
import numpy as np
import scipy.stats
from os import getcwd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from datetime import date
today = str(date.today())
```

```python
home = '/Users/lucindasisk/Dropbox/SNAPLab/Projects/2018_Cort_DTI/raw/'
dest = '/Users/lucindasisk/Dropbox/SNAPLab/Papers/D&P_2019/Figures'
data = pd.read_csv(home + '/ELST1_S_vals_forplot.csv').set_index('ID').dropna()
subjects = len(data)
datat = data.transpose()
```

```python
def return_vals(col, df):
    a = df.loc[col].as_matrix()
    x = a[~np.isnan(a)]
    stderr = scipy.stats.sem(x)
    mn = np.mean(x)
    return np.array([(mn - stderr), (mn + stderr), (mn)])

S1 = return_vals('S1', datat)
S2 = return_vals('S2', datat)
S3 = return_vals('S3', datat)
S4 = return_vals('S4', datat)
S5 = return_vals('S5', datat)

plot = np.array([S1, S2, S3, S4, S5]) # format: min, max, mean

line1 = plot[:,0] #min
line2 = plot[:,1] #max
line3 = plot[:,2] #mean
```

```python
fig, ax = plt.subplots()

x = [1,2,3,4,5]
ax.plot(x, line3, color = '#191970')

ax.set(xlabel='Collection Timepoint; N = {}'.format(subjects), ylabel='Cortisol Level (mg/dL)',
       title='Average Cortisol Levels During TSST')
ax.grid(axis='x')
ax.fill_between(x,line1,line2, color='#6495ED', alpha='0.5')
ax.fill_between(x,0,line1, color='#bdd3ee', alpha='0.5')

line_patch = mpatches.Patch(color='#191970', label='Mean Cortisol Trajectory')
err_patch = mpatches.Patch(color='#6495ED', label='Standard Error')
auc_patch = mpatches.Patch(color='#bdd3ee', label='Area Under the Curve (AUC)')
plt.legend(handles=[line_patch, err_patch, auc_patch], prop={'size': 8})


plt.xticks(np.arange(min(x), max(x)+1, 1.0))

fig.savefig(dest + "/Cortisol_Mean_Plot_"+today+".png")
plt.show()
```
